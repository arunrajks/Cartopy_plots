```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# seoul bike usage data

```python
df = pd.read_csv(
    "data/SeoulBikeData.csv"
)

df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df = df.dropna(axis=1)
# df.columns = colum_names


df
```

```python
plt.subplots(figsize=(14, 14))
i = 1
for label in df.columns[1:]:
    plt.subplot(4, 3, i)
    plt.scatter(df[label], df[df.columns[0]])
    plt.ylabel("bilke_count")
    plt.xlabel(label)
    i += 1

plt.tight_layout()
```

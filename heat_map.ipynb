```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "data/SeoulBikeData.csv"
)

colum_names = [
    "bike_count",
    "hour",
    "temp",
    "humidity",
    "wind",
    "visibility",
    "dew_pt_temp",
    "radiation",
    "rain",
    "snow",
    "functional",
]


df = df.drop(columns=["Date", "Holiday", "Seasons"])
df.columns = colum_names
df["functional"] = (df["functional"] == "Yes").astype(int)

df = df[df["hour"] == 12]
df = df.drop(["hour"], axis=1)
df=df.reset_index()
df
```

```python

plt.figure(figsize=(14,7))

# Add title
plt.title(" bike usage")

# Heatmap showing correlation map of bike usage
sns.heatmap(data=df[colum_names[0:1] + colum_names[2:-2]].corr(), annot=True)

```
